```python
import polars as pl
```

```python
samples = pl.read_parquet("../data/samples.parquet")
labels = pl.read_parquet("../data/labels.parquet")
pixel_data = pl.read_parquet("../data/pixel_data.parquet")
```

## Check for missing labels

```python
set(samples["sample_id"]) - set(labels["sample_id"])
```

## Check for missing pixel data

```python
set(samples["sample_id"]) - set(pixel_data["sample_id"])
```

## Check for incomplete time-series

```python
pixel_data.select("sample_id", "timestamps").group_by("sample_id").agg(
    pl.len(), pl.min("timestamps").alias("min"), pl.max("timestamps").alias("max")
).sort("len")
```
